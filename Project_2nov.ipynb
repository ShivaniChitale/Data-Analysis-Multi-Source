```python
products=['milk','tea','bread']
quanity_in_stock=[50,20,30]
import matplotlib.pyplot as plt
```

```python
explode=(0.05,0.05,0.05)
plt.pie(quanity_in_stock,labels=products,autopct='%1.1f%%',pctdistance=0.85,explode=explode)
center_circle=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(center_circle)
plt.title('products details')
plt.show()
```
